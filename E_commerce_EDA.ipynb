```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
plt.figure(figsize= (20,10))
sns.histplot(full_data_1['Prod_category'])
plt.xticks(rotation = 90)
plt.show()
```

```python
df_Ratings= full_data_1.sort_values(by='Ratings', ascending=False).head(10)
df_Ratings
```

```python
df_Discount= full_data_1.sort_values(by='Discount', ascending=False).head(10)
df_Discount
```

```python
col=['Ratings']
for i in col:
    sns.countplot(x= 'Prod_category', data = df_Ratings)
    plt.xticks(rotation = 90)
    plt.show()
```

```python
col=['Discount']
for i in col:
    sns.countplot(x= 'Prod_category', data = df_Discount)
    plt.figure()
```
